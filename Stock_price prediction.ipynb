```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplcursors
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from scipy.stats import norm
from IPython.display import display, clear_output
import time

# Function to fetch PE ratio
def fetch_pe_ratio(symbol):
    stock = yf.Ticker(symbol)
    pe_ratio = stock.info['trailingPE']
    return pe_ratio

# Function to calculate option Greeks
def calculate_greeks(S, K, T, r, sigma, option_type='call'):
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    if option_type == 'call':
        delta = norm.cdf(d1)
        gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))
        theta = -(S * norm.pdf(d1) * sigma) / (2 * np.sqrt(T)) - r * K * np.exp(-r * T) * norm.cdf(d2)
        vega = S * norm.pdf(d1) * np.sqrt(T)
        rho = K * T * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        delta = norm.cdf(d1) - 1
        gamma = norm.pdf(d1) / (S * sigma * np.sqrt(T))
        theta = -(S * norm.pdf(d1) * sigma) / (2 * np.sqrt(T)) + r * K * np.exp(-r * T) * norm.cdf(-d2)
        vega = S * norm.pdf(d1) * np.sqrt(T)
        rho = -K * T * np.exp(-r * T) * norm.cdf(-d2)
    return delta, gamma, theta, vega, rho

# Function to calculate RSI
def calculate_rsi(series, period):
    delta = series.diff(1)
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=period).mean()
    avg_loss = loss.rolling(window=period).mean()
    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Function to calculate additional indicators
def calculate_indicators(data):
    data['MA20'] = data['Close'].rolling(window=20).mean()
    data['BB_upper'] = data['MA20'] + 2 * data['Close'].rolling(window=20).std()
    data['BB_lower'] = data['MA20'] - 2 * data['Close'].rolling(window=20).std()
    data['RSI'] = calculate_rsi(data['Close'], 14)
    return data

# Function to calculate additional features
def calculate_features(data):
    data['Returns'] = data['Close'].pct_change()
    data['Log Returns'] = np.log(1 + data['Returns'])
    data['Volatility'] = data['Log Returns'].rolling(window=252).std() * np.sqrt(252)
    data['Implied Volatility'] = data['Volatility']
    
    K = data['Close']
    T = 30 / 365
    r = 0.01
    
    greeks = data.apply(lambda row: calculate_greeks(row['Close'], row['Close'], T, r, row['Implied Volatility']), axis=1)
    greeks_df = pd.DataFrame(greeks.tolist(), index=data.index, columns=['Delta', 'Gamma', 'Theta', 'Vega', 'Rho'])
    data = pd.concat([data, greeks_df], axis=1)
    
    data = calculate_indicators(data)
    
    data['Buy_Signal'] = (data['Close'] < data['BB_lower']) & (data['RSI'] < 30)
    data['Sell_Signal'] = (data['Close'] > data['BB_upper']) & (data['RSI'] > 70)
    
    return data

# Function to download data
def download_data(stock_symbols, start_date='2020-01-01'):
    stock_data = {}
    for symbol in stock_symbols:
        data = yf.download(symbol, start=start_date)
        stock_data[symbol] = data
        print(f"Downloaded data for {symbol}")
    return stock_data

# Function to fetch the latest data
def fetch_latest_data(symbol, start_date='2020-01-01'):
    return yf.download(symbol, start=start_date)

# List of stock symbols
stocks = ['TSLA', 'AAPL', 'NFLX', 'ABNB', 'PLTR', 'NVDA']

# Download initial data for the specified stocks
stock_data = download_data(stocks)

# Display the first few rows of the data for each stock
for symbol in stocks:
    print(f"Data for {symbol}:")
    print(stock_data[symbol].head())
    print("\n")

# Calculate features for each stock
for symbol in stocks:
    print(f"Calculating features for {symbol}...")
    stock_data[symbol] = calculate_features(stock_data[symbol])

# Display the first few rows of the data with new features for each stock
for symbol in stocks:
    print(f"Data for {symbol} with features:")
    print(stock_data[symbol].head())
    print("\n")

# Prepare data for LSTM
def prepare_data_with_pe(data, training_ratio=0.8):
    df = data[['Close', 'Implied Volatility', 'Delta', 'Gamma', 'Theta', 'Vega', 'Rho', 'MA20', 'BB_upper', 'BB_lower', 'RSI', 'Buy_Signal', 'Sell_Signal']].dropna()

    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df)

    training_data_len = int(np.ceil(len(scaled_data) * training_ratio))

    train_data = scaled_data[0:training_data_len]
    test_data = scaled_data[training_data_len - 60:]

    x_train, y_train = [], []
    for i in range(60, len(train_data)):
        x_train.append(train_data[i-60:i])
        y_train.append(train_data[i, 0])  # Predict 'Close' price

    x_train, y_train = np.array(x_train), np.array(y_train)

    x_test, y_test = [], df[training_data_len:]['Close'].values
    for i in range(60, len(test_data)):
        x_test.append(test_data[i-60:i])

    x_test = np.array(x_test)

    return x_train, y_train, x_test, y_test, scaler

# Prepare the data for each stock
prepared_data = {}
for symbol in stocks:
    print(f"Preparing data for {symbol} with PE ratio...")
    x_train, y_train, x_test, y_test, scaler = prepare_data_with_pe(stock_data[symbol])
    prepared_data[symbol] = {
        'x_train': x_train,
        'y_train': y_train,
        'x_test': x_test,
        'y_test': y_test,
        'scaler': scaler
    }
    print(f"Shapes for {symbol}:")
    print(f"x_train: {prepared_data[symbol]['x_train'].shape}")
    print(f"y_train: {prepared_data[symbol]['y_train'].shape}")
    print(f"x_test: {prepared_data[symbol]['x_test'].shape}")
    print(f"y_test: {prepared_data[symbol]['y_test'].shape}")
    print("\n")

# Build and train LSTM model
def build_and_train_model(x_train, y_train, epochs=1, batch_size=1):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dense(units=25))
    model.add(Dense(units=1))

    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs)

    return model

# Train the model for each stock
models = {}
for symbol in stocks:
    print(f"Training model for {symbol}...")
    data = prepared_data[symbol]
    model = build_and_train_model(data['x_train'], data['y_train'])
    models[symbol] = model

    # Save the model for future use
    model.save(f'{symbol.lower()}_model.h5')
    print(f"Model saved for {symbol}")

# Confirm that models are trained and saved
for symbol in stocks:
    print(f"Model trained and saved for {symbol}")

# Make predictions
def make_predictions(model, x_test, y_test, scaler):
    predictions = model.predict(x_test)
    predictions = scaler.inverse_transform(np.concatenate((predictions, np.zeros((predictions.shape[0], x_test.shape[2] - 1))), axis=1))[:, 0]

    rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
    return predictions, rmse

# Make predictions for each stock
predictions = {}
for symbol in stocks:
    print(f"Making predictions for {symbol}...")
    data = prepared_data[symbol]
    model = models[symbol]
    preds, rmse = make_predictions(model, data['x_test'], data['y_test'], data['scaler'])
    predictions[symbol] = preds
    print(f"{symbol} RMSE: {rmse}")

# Plot predictions with improved visualization
import mplcursors

# Plot predictions with improved visualization
def plot_predictions(data, predictions, title):
    train = data[:len(data) - len(predictions)]
    valid = data[len(data) - len(predictions):]
    valid['Predictions'] = predictions

    plt.figure(figsize=(16, 8))
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price USD ($)')
    plt.plot(train['Close'], label='Train')
    plt.plot(valid[['Close']], label='Validation')
    plt.plot(valid[['Predictions']], label='Predictions')
    plt.plot(data['MA20'], label='20-Day MA', linestyle='--')
    plt.plot(data['BB_upper'], label='Upper Bollinger Band', linestyle='--')
    plt.plot(data['BB_lower'], label='Lower Bollinger Band', linestyle='--')

    # Plot buy/sell signals with a minimum distance apart to avoid clutter
    buy_signals = data[data['Buy_Signal']].index
    sell_signals = data[data['Sell_Signal']].index

    min_dist = 10  # Minimum distance between signals
    filtered_buy_signals = [buy_signals[i] for i in range(len(buy_signals)) if i == 0 or (buy_signals[i] - buy_signals[i - 1]).days > min_dist]
    filtered_sell_signals = [sell_signals[i] for i in range(len(sell_signals)) if i == 0 or (sell_signals[i] - sell_signals[i - 1]).days > min_dist]

    plt.scatter(filtered_buy_signals, data.loc[filtered_buy_signals]['Close'], label='Buy Signal', marker='^', color='green')
    plt.scatter(filtered_sell_signals, data.loc[filtered_sell_signals]['Close'], label='Sell Signal', marker='v', color='red')

    # Add mplcursors for interactive annotations
    cursor = mplcursors.cursor(hover=True)
    cursor.connect("add", lambda sel: sel.annotation.set_text(
        f'Date: {data.index[sel.index].date()}\nPrice: {sel.target[1]:.2f}'))

    plt.legend(loc='lower right')
    plt.show()

# Example usage
# Plot predictions for each stock
for symbol in stocks:
    print(f"Plotting predictions for {symbol}...")
    plot_predictions(stock_data[symbol], predictions[symbol], f'{symbol} Stock Price Prediction')


# Continuous monitoring with live plotting
def continuous_monitoring(stocks, interval=3600):
    latest_data = {symbol: stock_data[symbol] for symbol in stocks}
    
    plt.ion()
    fig, axs = plt.subplots(len(stocks), 1, figsize=(16, 8 * len(stocks)))

    while True:
        for i, symbol in enumerate(stocks):
            print(f"Fetching latest data for {symbol}...")
            new_data = fetch_latest_data(symbol)
            
            # Update the stored data with new data
            latest_data[symbol] = new_data
            
            # Calculate features
            updated_data = calculate_features(latest_data[symbol])
            
            # Ensure all necessary columns are present
            required_columns = ['MA20', 'BB_upper', 'BB_lower', 'RSI']
            if not all(col in updated_data.columns for col in required_columns):
                print(f"Error: Missing indicators for {symbol}. Skipping...")
                continue
            
            # Prepare the data
            x_train, y_train, x_test, y_test, scaler = prepare_data_with_pe(updated_data)
            
            # Train or update the model
            model = build_and_train_model(x_train, y_train, epochs=1, batch_size=1)
            
            # Make predictions
            predictions, rmse = make_predictions(model, x_test, y_test, scaler)
            print(f"{symbol} RMSE: {rmse}")
            
            # Plot predictions
            axs[i].clear()
            axs[i].set_title(f'{symbol} Stock Price Prediction')
            axs[i].set_xlabel('Date')
            axs[i].set_ylabel('Close Price USD ($)')
            train = latest_data[symbol][:len(latest_data[symbol]) - len(predictions)]
            valid = latest_data[symbol][len(latest_data[symbol]) - len(predictions):]
            valid['Predictions'] = predictions
            axs[i].plot(train['Close'], label='Train')
            axs[i].plot(valid[['Close']], label='Validation')
            axs[i].plot(valid[['Predictions']], label='Predictions')
            axs[i].plot(updated_data['MA20'], label='20-Day MA', linestyle='--')
            axs[i].plot(updated_data['BB_upper'], label='Upper Bollinger Band', linestyle='--')
            axs[i].plot(updated_data['BB_lower'], label='Lower Bollinger Band', linestyle='--')
            
            # Plot buy/sell signals with a minimum distance apart to avoid clutter
            buy_signals = updated_data[updated_data['Buy_Signal']].index
            sell_signals = updated_data[updated_data['Sell_Signal']].index

            min_dist = 10  # Minimum distance between signals
            filtered_buy_signals = [buy_signals[i] for i in range(len(buy_signals)) if i == 0 or (buy_signals[i] - buy_signals[i - 1]).days > min_dist]
            filtered_sell_signals = [sell_signals[i] for i in range(len(sell_signals)) if i == 0 or (sell_signals[i] - sell_signals[i - 1]).days > min_dist]

            axs[i].scatter(filtered_buy_signals, updated_data.loc[filtered_buy_signals]['Close'], label='Buy Signal', marker='^', color='green')
            axs[i].scatter(filtered_sell_signals, updated_data.loc[filtered_sell_signals]['Close'], label='Sell Signal', marker='v', color='red')

            # Add mplcursors for interactive annotations
            cursor = mplcursors.cursor(hover=True)
            cursor.connect("add", lambda sel: sel.annotation.set_text(
                f'Date: {updated_data.index[sel.index].date()}\nPrice: {sel.target[1]:.2f}'))

            axs[i].legend(loc='lower right')
        
        display(fig)
        plt.pause(interval)
        clear_output(wait=True)

# List of stock symbols
stocks = ['TSLA', 'AAPL', 'NFLX', 'ABNB', 'PLTR', 'NVDA']

# Start continuous monitoring
continuous_monitoring(stocks, interval=3600)  # Adjust the interval as needed
```
